## 1 : Basic Class and OObject

## 3 : Inheritance

```python
                    # This is Parent Class

class Car:
    def __init__(self,brand,model):             # Constructor method
        self.brand = brand                           # Initialize the brand attribute for the object.
        self.model = model                         # Initialize the model attribute for the object.
        
    def full_name(self):                           # Defining a method named full_name
        return f'{self.brand} {self.model}'         # Returning a formatted string with brand and model
    
    
                    # This is Child Class
                    
class electricCar(Car):                                             # Defining a class named electricCar that inherits from Car
    def __init__(self, brand, model, battery_size):          # Constructor method for child class
        super().__init__(brand, model)                     # Calling the constructor of the parent class to initialize brand and model
        self.battery_size = battery_size                 # Initialize the battery_size attribute for the object.
        
e1 = electricCar('tesla', 'model s', 100)         # Creating an object of the electricCar class
print(e1.full_name())                       # Calling the full_name method inherited from the Car class
print(e1.battery_size)                     # Accessing the battery_size attribute of the electricCar object
```

## 4 : Encapsulation

Encapsulation means hiding the internal details of a class and protecting the data from direct access.
In Python, we do this by:
- Using public attributes → can be accessed anywhere
- Using protected attributes (_attribute) → should not be accessed directly
- Using private attributes (__attribute) → cannot be accessed directly
- Using methods to control how data is accessed or changed
Encapsulation = data protection + controlled access


```python
class Person():
    def __init__(self, name):
        self.name = name        #this is the public attribute
        
p = Person('waqar')
print(p.name)           # public attribute accessible everywhere
```

#### 4.2  Protected Attributes

```python
class person:
    def __init__(self, name):
        self._name = name      #protected attribute
        
    
p = person('waqar')
print(p._name)          # protected attribute accessible within class and subclasses and by convention should not be accessed directly outside the class
```

### 4.3  Private Attributes

```python
class Person:
    def __init__(self, name):
        self.__name = name     # private attribute
    
p = Person('waqar')
print(p.__name)         # private attribute not accessible outside the class directly
```

### 4.4 Accessing Private Attributes Using Methods

```python
class Person:
    def __init__(self, name):
        self.__name = name     # private attribute
        
    def get_name(self):
        return self.__name    # public method to access private attribute
    
    def set_name(self,new_name):
        self.__name = new_name  # public method to set private attribute
        
p = Person('waqar')
print(p.get_name())         # accessing private attribute via public method
p.set_name('khan')          # modifying private attribute via public method
print(p.get_name())         # accessing modified private attribute via public method
```

###  4.5 Full Example (Simple and Clear)

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance                # private attribute

    def get_balance(self):              # public method to access private attribute
        return self.__balance           # returning the balance
    
    def deposit(self,  amount):         # public method to modify private attribute
        if amount > 0:             # checking for valid amount
            self.__balance +=amount     # updating the balance
            return 'deposit successful'  # returning success message
        return 'invalid amount'    # returning error message for invalid amount
    
a = BankAccount(1000)       # creating object of BankAccount class
print(a.get_balance())      # accessing balance via public method
print(a.deposit(500))    # depositing amount via public method
print(a.get_balance())      # accessing updated balance via public method
```

# 5. Polymorphism

```python
class Dog:                      # defining Dog class
    def sound(self):           # defining sound method
        return 'woof!'      # returning sound of dog
    
class Cat():                     # defining Cat class
    def sound(self):       # defining sound method
        return 'Meow!'    # returning sound of cat
    
d = Dog()                # creating object of Dog class
c = Cat()       

for animal in (d,c):      # iterating over dog and cat objects
    print(animal.sound())   # calling sound method for each object and each method behaves differently
```

```python
    # polymorphism means the same name of methods in different classes with different behaviors when called on their respective objects

class Bird:                 # defining Bird class
    def fly(self):               # defining fly method
        return "Bird is flying"             # returning flying message

class Sparrow(Bird):      # defining Sparrow class that inherits from Bird class
    def fly(self):                   # overriding fly method
        return "Sparrow flies fast"     # returning specific flying message for Sparrow

class Penguin(Bird):                        # defining Penguin class that inherits from Bird class
    def fly(self):                 # overriding fly method
        return 'Penguin cannot fly'      # returning specific flying message for Penguin
    
b1 = Sparrow()               # creating object of Sparrow class
b2 = Penguin()         # creating object of Penguin class
print(b1.fly())          # calling fly method for Sparrow object
print(b2.fly())     # calling fly method for Penguin object
```

## Exercise

Parent Class: Shape

- Method: area() → returns "Area not defined"

Child Class: Circle

- Overrides area()
- Returns: "Circle area = <value>"
(You can use a simple formula: 3.14 * r * r)

Child Class: Square

- Overrides area()
- Returns: "Square area = <value>"
(Use formula: side * side)

Then:

- Create one Circle object
- Create one Square object
- Call area() for both

```python
class Shape:
    def area(self):
        return "Area not defined"
    
class Circle(Shape):
    def __init__(self, circle_area):
        self.circle_area = circle_area
        
    def area(self):
        return f'the area of the circle is {3.14 * self.circle_area * self.circle_area}'

class Square(Shape):
    def __init__(self,square_area):
        self.square_area = square_area
    
    def area(self):
        return f'the area of the square is {self.square_area * self.square_area}'

c = Circle(5)
s = Square(4)
print(c.area())
print(s.area())
```

## 7 : Static Method

Use a static method when:
- You want a function inside a class
- But the function does not need object attributes
- And does not need class variables
- we can call the the static method through class name and through object as well.

Before the Satatic method 
- we should use @staticmethod


```python
class Mathtools:               # defining Mathtools class
    
    @staticmethod              # defining static method
    def add(a,b):              # defining add method
        return a+b              # returning sum of a and b

mt = Mathtools()               # creating object of Mathtools class but its not necessary to create object to call static method
print(Mathtools.add(2,5))       # calling static method using class name
print(mt.add(3,4))              # calling static method using object
```

```python
class Employee:          # defining Employee class
    company  = 'microsoft'        # class attribute shared by all instances
    
    
    def __init__(self, name):         # constructor method
        self.name = name       # instance attribute
        
    @staticmethod         # defining static method
    def is_valid_name(name):    # static method to validate name
        return name.isalpha()   # returning True if name contains only alphabets, else False
    
print(Employee.is_valid_name('waqar'))  # calling static method using class name
print(Employee.is_valid_name('khan123'))    # calling static method using class name
```

## Exercise

Create a class Calculator with:
Static methods:

- add(a, b)
- subtract(a, b)
- multiply(a, b)
- divide(a, b) → return "Cannot divide by zero" if b == 0

Then:

- Call each method using the class name
- Print the results


```python
class Calculator:
    
    @staticmethod
    def add(a,b):
        c = a+b
        return f'the sum of {a} and {b} is {c}'
    
    @staticmethod
    def sub(a,b):
        c = a - b
        return f'the subtraction of {a} from {b} is {c}'
    
    @staticmethod
    def mul(a,b):
        c = a * b
        return f'the multiplication of {a}, and {b} is {c}'
    
    @staticmethod
    def div(a,b):
        if b == 0:
            return 'division by zero is not allowed'
        else:
            c = a / b
            return f'the division of {a} by {b} is {c}'

print(Calculator.add(5,7))
print(Calculator.sub(10,3))
print(Calculator.mul(4,6))
print(Calculator.div(8,0))
```

## 8 : Property Decorators

Property Decorators means we can use getters and setters in python using property decorators

```python
class BankAccount:         # defining BankAccount class
    def __init__(self, balance):        # constructor method
        self.__balance = balance        # private attribute
     
    @property        # getter method
    def balance(self):          # defining balance method
        return self.__balance        # returning the balance
    @balance.setter        # setter method
    def balance(self,amount):         # defining balance method to set balance
        if amount >= 0:       # checking for valid amount
            self.__balance = amount     # updating the balance
            
        else:                       
            print('invalid amount')     # printing error message for invalid amount

a = BankAccount(1000)       # creating object of BankAccount class
print(a.balance)            # accessing balance via getter method
a.balance = 1500            # setting balance via setter method
print(a.balance)            # accessing updated balance via getter method
            
```

## Exercise

Create a class Student with:
Private attribute:

- __grade

Property methods:

- @property grade → returns grade
- @grade.setter → only allow grades between 0 and 100

Then:

- Create a student
- Print the grade
- Update the grade
- Try setting an invalid grade


```python
class Student:
    def __init__(self,grade):
        self.__grade = grade
        
    @property
    def grade(self):
        return self.__grade
    
    @grade.setter
    def grade(self, new_grade):
        if new_grade >=0 and new_grade <=100:
            self.__grade = new_grade
        else:
            print('invalid grade')

s = Student(94)
print(s.grade)
s.grade = 105
print(s.grade)
```

## 10 : Multiple Inheritance

- A class can inherit from more than one parent class.

Method Overriding in Multiple Inheritance

- If both parents have the same method name, Python uses MRO (Method Resolution Order) to decide which one to call.

```python
class A:
    def featureA(self):
        return 'Feature A'
    
class B:
    def featureB(self):
        return 'feature B'
    
class C(A, B):
    pass


c = C()
print(c.featureA())
print(c.featureB())
```

```python
# Method Overriding in Multiple Inheritanc 

#  Python uses MRO (Method Resolution Order) to decide which one to call.


class A:
    def show(self):
        return "A"

class B:
    def show(self):
        return "B"

class C(A, B):
    pass

c = C()
print(c.show())  # It will check class C first. If C does not have the method, then it checks the parent classes in order. When it finds the method in one of the parents, it stops and displays that method
print(C.mro())   # shows the exact order in which Python will search for methods when you call them on an object of class C.
                 # It starts with class C, then checks its parent classes from left to right, and finally checks the base class object.
```

```python
class Father:
    def skills(self):
        return "Driving, Repairing"
    
class Mother:
    def skills(self):
        return "cooling, teaching"
    
class Child(Father,Mother):
    pass

c = Child()
print(c.skills())
```
