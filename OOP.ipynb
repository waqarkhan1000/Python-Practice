## 1 : Basic Class and OObject

```python
class Car:                                                  # Defining a class named Car
    def __init__(self, make, model, brand):                 # Constructor method
        self.make = make                                    # Initialize the make attribute for the object.
        self.model = model
        self.brand = brand
        
    # method of the class
    def high_speed(self, speed):                           # Defining a method named high_speed    
        return f'The {self.make} {self.model} is running at high speed {speed} km/h'  # Returning a formatted string with make, model, and speed

c1 = Car('toyota', 2025, 'corolla')                    # Creating an object of the car class    
print(c1.make)                                     # Accessing the make attribute of the object 
print(c1.high_speed('120'))                # Calling the high_speed method of the object
```

## Exercise
Create a class named Student with:
- a constructor that takes name and age
- attributes to store both
- a method called info() that prints:
"Student Name: <name>, Age: <age>"
Then create two objects of Student and call the method.


```python
class Student():
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def info(self):
        return f'the name of the student is {self.name} and age is {self.age}'

s1 = Student('waqar', 22)
s2 = Student('khan', 25)
print(s1.info())
print(s2.info())
```

## 2 : Class Method and Self

```python
class Car():
    def __init__(self, brand, model):                   # without self object cannot store and access attributes
        self.brand = brand
        self.model = model
    
    def full_name(self):
        return f"{self.brand} {self.model}"
    
c1 = Car('toyota', 'colrola')
c2 = Car("audi ", 'A4')
print(c1.full_name())
print(c2.full_name())
```

## Exercise

Create a class Laptop with:
- Constructor that takes brand and ram
- A method details() that returns:
"Laptop Brand: <brand>, RAM: <ram>GB"
Then:
- Create two objects
- Call the method for both objects
- Print the results


```python
class Laptop():
    def __init__(self, brand, ram):
        self.brand = brand
        self.ram = ram
        
    def details(self):
        return f"the {self.brand} laptop has {self.ram}GB of RAM"
    
l1 = Laptop('dell', 16)
l2 = Laptop('hp', 8)

print(l1.details())
print(l2.details())
```
