## 1 : Basic Class and OObject

```python
class Car:                                                  # Defining a class named Car
    def __init__(self, make, model, brand):                 # Constructor method
        self.make = make                                    # Initialize the make attribute for the object.
        self.model = model
        self.brand = brand
        
    # method of the class
    def high_speed(self, speed):                           # Defining a method named high_speed    
        return f'The {self.make} {self.model} is running at high speed {speed} km/h'  # Returning a formatted string with make, model, and speed

c1 = Car('toyota', 2025, 'corolla')                    # Creating an object of the car class    
print(c1.make)                                     # Accessing the make attribute of the object 
print(c1.high_speed('120'))                # Calling the high_speed method of the object
```

## 3 : Inheritance

```python
                    # This is Parent Class

class Car:
    def __init__(self,brand,model):             # Constructor method
        self.brand = brand                           # Initialize the brand attribute for the object.
        self.model = model                         # Initialize the model attribute for the object.
        
    def full_name(self):                           # Defining a method named full_name
        return f'{self.brand} {self.model}'         # Returning a formatted string with brand and model
    
    
                    # This is Child Class
                    
class electricCar(Car):                                             # Defining a class named electricCar that inherits from Car
    def __init__(self, brand, model, battery_size):          # Constructor method for child class
        super().__init__(brand, model)                     # Calling the constructor of the parent class to initialize brand and model
        self.battery_size = battery_size                 # Initialize the battery_size attribute for the object.
        
e1 = electricCar('tesla', 'model s', 100)         # Creating an object of the electricCar class
print(e1.full_name())                       # Calling the full_name method inherited from the Car class
print(e1.battery_size)                     # Accessing the battery_size attribute of the electricCar object
```

## Exercise

Parent Class: Animal
- attributes: name, age
- method: info() → returns "Animal: <name>, Age: <age>"

Child Class: Dog
- inherits from Animal
- adds attribute: breed
- method: sound() → returns "Woof!"

Then:
- Create one Dog object
- Print:
- info() (inherited)
- sound() (child method)
- breed (child attribute)


```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def info(self):
        return f'The name of the animal is {self.name} and age is {self.age}'

class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)
        self.breed = breed
        
    def sound(self):
        return f'the {self.name} barks woof woof'
    
d1 = Dog('buddy', 3, 'golden retriever')
print(d1.info())
print(d1.sound())
print(d1.breed)
```

1. Create a Parent Class: Vehicle
- Attributes:
- brand
- model
- Method:
- info() → returns "Brand: <brand>, Model: <model>"

2. Create a Child Class: Car (inherits from Vehicle)
- Additional attribute:
- seats
- Additional method:
- car_details() → returns "This car has <seats> seats"

3. Create a Child Class: Bike (inherits from Vehicle)
- Additional attribute:
- cc (engine power)
- Additional method:
- bike_details() → returns "This bike has <cc>cc engine"

4. Create Objects
- Create one Car object
- Create one Bike object

5. Print the Following for Each Object
- info() → inherited from Vehicle
- Child‑specific method:
- car_details() for Car
- bike_details() for Bike
- Child‑specific attribute:
- seats for Car
- cc for Bike


```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def info(self):
        return f'The vehicle is a {self.brand} {self.model}'
    
class Car(Vehicle):
    def __init__(self, brand, model, seats):
        super().__init__(brand, model)
        self.seats = seats
    
    def car_details(self):
        return f'the car has {self.seats} seats.'
    
class Bike(Vehicle):
    def __init__(self, brand, model, cc):
        super().__init__(brand, model)
        self.cc = cc
        
    def bike_details(self):
        return f'the bike has {self.cc} cc engine.'
    
car = Car('honda', 'civic', 5)
bik = Bike('yamaha', 'r15', 150)
print(car.info())
print(bik.info())
print(car.car_details())
print(car.seats)
print(bik.bike_details())
print(bik.cc)
```

## 4 : Encapsulation

Encapsulation means hiding the internal details of a class and protecting the data from direct access.
In Python, we do this by:
- Using public attributes → can be accessed anywhere
- Using protected attributes (_attribute) → should not be accessed directly
- Using private attributes (__attribute) → cannot be accessed directly
- Using methods to control how data is accessed or changed
Encapsulation = data protection + controlled access


```python
class Person():
    def __init__(self, name):
        self.name = name        #this is the public attribute
        
p = Person('waqar')
print(p.name)           # public attribute accessible everywhere
```

#### 4.2  Protected Attributes

```python
class person:
    def __init__(self, name):
        self._name = name      #protected attribute
        
    
p = person('waqar')
print(p._name)          # protected attribute accessible within class and subclasses and by convention should not be accessed directly outside the class
```

### 4.3  Private Attributes

```python
class Person:
    def __init__(self, name):
        self.__name = name     # private attribute
    
p = Person('waqar')
print(p.__name)         # private attribute not accessible outside the class directly
```
